```python
def get_store_list_per_concept(country):
    query     = '''select * from Tier_parameters where country = "{0}"
                   and concept in ("TIER_SULTAN","TIER_MARKET","TIER_ONLINE","TIER_EXPRESS","TIER_WAREHOUSE","TIER_COMMUNITY","TIER_PROJECT") 
                   and tier = "T1" 
                '''.format(country)
    store     = pd.read_sql(query,engine)

    a=[]
    b=[]
    c=[]
    
    for concept in store['concept']:
        store_list = list(eval(store[store['concept']==concept].reset_index()['store_list'].iloc[0]))
        print(store_list)
        a.extend(store_list)
        b.extend([country]*len(store_list))
        c.extend([concept]*len(store_list))

    final = pd.DataFrame(columns=['country','concept','store_list'])
    final['country']    = b
    final['concept']    = c
    final['store_list'] = a
    
    return final
```
